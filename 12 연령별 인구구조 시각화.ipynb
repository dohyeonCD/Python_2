# 시도-연령별 인구구조

## 서브셋 만들기
* `컬럼.str.contains(문자열)` : 해당 문자열이 포함된 데이터 추출

```python
df_sido= df[df.index.str.contains('00000000')]
df_sido.head(2)
```

## 시각화

### 다중플롯

```python
plt.rcParams['figure.figsize']=(12,5)

for i in range(len(df_sido)):
    plt.plot(df_sido.iloc[i], label=df_sido.index[i].split(' ')[0])


plt.xticks(range(0,101,5))
plt.xlabel('연령')
plt.ylabel('인구수')
plt.title('2021년 10월 시도- 연령별 인구수', size=20)
plt.legend(loc=(1.01,0.06))
plt.show()
```

### 서브플롯
* `plt.subplot(행,열,순번)`

```python
fig= plt.figure(figsize=(20,30))

for i in range(len(df_sido)):
    plt.subplot(6,3,i+1)
    plt.plot(df_sido.iloc[i])
    plt.xticks(range(0,101,10))
    plt.title(df_sido.index[i].split(' ')[0])
    plt.grid(':', alpha=0.5)
    
fig.tight_layout()    
plt.show()
```

# 특정 동 인구구조

## 학군지 인구구조
* 대치, 목
* `fig, ax = plt.subplots()`

```python
df_dong= df[(df.index.str.contains('강남구 대치'))|(df.index.str.contains('양천구 목'))].copy()
```

```python
fig, ax= plt.subplots(3,3, figsize=(15,10), sharey=True)
r,c= 0,0 #서브 플롯의 행번호는 r, 열번호는 c. 초기값 지정해줌.

for i in range(len(df_dong)):
    ax[r][c].plot(df_dong.iloc[i])
    ax[r][c].set_xticks(range(0,101,10), rotation=90)
    ax[r][c].set_title(df_dong.index[i].split(' ')[2].split('(')[0])
    
    c+=1
    if c%3==0:
        r+=1
        c=0
    
fig.tight_layout()    
plt.show()
```
